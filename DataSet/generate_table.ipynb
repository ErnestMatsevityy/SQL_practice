```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Задаємо дати початку і закінчення
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 6, 30)

# Задаємо кількість унікальних користувачів
num_users = 4200

# Список країн Європи
european_countries = ['Austria', 'Albania', 'Andorra', 'Belgium', 'Bulgaria', 'Bosnia and Herzegovina', 'United Kingdom', 'Greece', 'Denmark', 'Estonia', 'Ireland', 'Iceland', 'Spain', 'Italy', 'Kosovo', 'Latvia', 'Lithuania', 'Liechtenstein', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Netherlands', 'Germany', 'Norway', 'North Macedonia', 'Poland', 'Portugal', 'Romania', 'San Marino', 'Serbia', 'Slovakia', 'Slovenia', 'Ukraine', 'Finland', 'France', 'Croatia', 'Czech Republic', 'Montenegro', 'Switzerland', 'Sweden']

# Генеруємо дані для кожної колонки
data = {
    'user_id': np.arange(1, num_users + 1),
    'age': np.random.randint(16, 51, size=num_users),
    'gender': np.random.choice(['Male', 'Female'], size=num_users),
    'location': np.random.choice(european_countries, size=num_users),
    'time_spent': np.random.randint(1, 3600, size=num_users),  # в секундах
    'num_sessions': np.random.randint(1, 101, size=num_users),
    'level': np.random.randint(1, 41, size=num_users),
}

# Залежні колонки
data['tasks_completed'] = data['level'] * 5
data['registration_date'] = [start_date + timedelta(days=np.random.randint((end_date - start_date).days))
                              for _ in range(num_users)]

# Генеруємо user_event для кожного користувача
user_events = []
for user_id in data['user_id']:
    num_events = np.random.randint(1, 13)
    event_dates = [data['registration_date'][user_id - 1] + timedelta(days=np.random.randint((end_date - data['registration_date'][user_id - 1]).days))
                   for _ in range(num_events)]
    
    # Додаємо різні типи івентів (login, shop, view_advert) з відповідними умовами
    for event_date in event_dates:
        event_type = np.random.choice(['login', 'shop', 'view_advert'])
        purchases = np.random.randint(0, 101) if event_type == 'shop' else 0
        revenue = np.random.randint(0, 101) if event_type == 'shop' else 0
        user_events.append((user_id, event_date, revenue, purchases, event_type))

# Створюємо DataFrame для user_events
user_events_df = pd.DataFrame(user_events, columns=['user_id', 'event_date', 'revenue', 'purchases', 'event_type'])

# Додаємо колонку Earnings_advertisement
user_events_df['Earnings_advertisement'] = np.where((user_events_df['event_type'] == 'login') | (user_events_df['event_type'] == 'shop'), 0, np.random.randint(0, 101, size=len(user_events_df)))

# Об'єднуємо дані в один DataFrame
df = pd.DataFrame(data)
df = pd.merge(df, user_events_df, on='user_id', how='left')

df.to_csv('mobile_game_analytics_data_5.csv', index=False)
```
